Pie chart

```python
# Importing the pyplot module from matplotlib for plotting
import matplotlib.pyplot as plt

# Creating a new figure (a blank canvas for plots)
figure = plt.figure()

# Adding an axes (area to draw the chart) that takes up the full figure
# [0, 0, 1, 1] means: start from left=0, bottom=0, width=1 (100%), height=1 (100%)
ax = figure.add_axes([0, 0, 1, 1])

# List of languages to be used as labels in the pie chart
languages = ['English', 'Nepali', 'Spanish', 'Latin', 'Germany']

# Number of people who speak each corresponding language
people = [100, 50, 150, 40, 70]

# Creating a pie chart:
# - 'people' defines the sizes of the slices
# - 'labels=languages' assigns labels to each slice
# - 'autopct' shows percentage values on the slices, formatted to 1 decimal place
ax.pie(people, labels=languages, autopct='%1.1f%%')

# Display the final chart on the screen
plt.show()
```

Scatter plot

```python
# Importing necessary libraries
import numpy as np              # For numerical operations like creating arrays
import matplotlib.pyplot as plt # For plotting

# Generating 30 evenly spaced values between 0 and 10
x = np.linspace(0, 10, 30)

# Calculating sine values for each x
y = np.sin(x)

# Calculating cosine values for each x
z = np.cos(x)

# Creating a new figure (canvas for plotting)
fig2 = plt.figure()

# Adding a full-sized axes to the figure
# [0, 0, 1, 1] means it starts from bottom-left corner and takes full width and height
ax = fig2.add_axes([0, 0, 1, 1])

# Plotting sine values as green dots
ax.scatter(x, y, color='g')  # 'g' stands for green

# Plotting cosine values as blue dots
ax.scatter(x, z, color='b')  # 'b' stands for blue

# Displaying the scatter plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create time values
time = np.linspace(0, 10, 30)

# Spring A: Sine wave
y = np.sin(time)

# Spring B: Scaled cosine wave
z = 0.5 * np.cos(time)

# Create a figure and axis
figure1 = plt.figure()
ax = figure1.add_axes([0, 0, 1, 1])

# Scatter plot for Spring A
ax.scatter(time, y, color='r', label='Spring A')

# Scatter plot for Spring B
ax.scatter(time, z, color='purple', label='Spring B')

# Add labels and title
ax.set_xlabel("Time (seconds)")
ax.set_ylabel("Displacement (meters)")
ax.set_title("Spring Displacement Over Time")

# Add legend to identify the two springs
ax.legend()

# Show the plot
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure() #creating a new figure
ax = plt.axes(projection='3d') #Add a 3D axes to the figure using the 'Projection' argument
z= 20 * np.random.random(100) # Generate 100 random values between 0 and 20 for z-axis

#calculate x and y using sine and consine of -z values
x=np.sin(z)
y=np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Purples')
plt.show()
```
