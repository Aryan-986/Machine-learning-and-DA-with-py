```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Import seaborn and load the 'tips' dataset
tips = sns.load_dataset('tips')  # Loads built-in restaurant tips data from seaborn

# Create a relational scatter plot to visualize tipping patterns
sns.relplot(
    data=tips,             # Use the 'tips' DataFrame as the data source
    x='total_bill',        # x-axis: total amount of the bill
    y='tip',               # y-axis: tip given
    col='sex',             # Creates separate plots (columns) for 'Male' and 'Female' customers
    hue='day',             # Color points based on the day of the week (Thur, Fri, Sat, Sun)
    style='day',           # Use different marker shapes for different days
    size='size'            # Size of each point represents the number of people in the group
)
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load California housing dataset
house_data = fetch_california_housing()

# Create a DataFrame
house = pd.DataFrame(house_data.data, columns=house_data.feature_names)
house['PRICE'] = house_data.target

house.head()  # View the first few rows
```

Distribution plot

```python
sns.displot(house['PRICE'], kde=True)
```

Corelation:


*   Positive Correlation
*   Negative Correlatio



Heat Map

```python
correlation = house.corr()
```

```python
# Constructing a heatmap of correlations
plt.figure(figsize=(10, 10))  # Set the size of the entire figure

# Draw the heatmap
sns.heatmap(
    correlation,               # 'correlation' should be a DataFrame, typically from df.corr()
    cbar=True,                 # Show color bar
    square=True,               # Make cells square
    fmt='.1f',                 # Format the annotation to 1 decimal place
    annot=True,                # Annotate each cell with the numeric value
    annot_kws={'size': 8},     # Set the font size of annotations
    cmap='Blues'               # Use 'Blues' colormap
)
```
